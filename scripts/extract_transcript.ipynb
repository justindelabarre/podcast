```python
# ─── Configuration ───────────────────────────────────────────────
# Modifier ces deux variables avant chaque transcription
EPISODE_SLUG = "blef_fivetran_x_dbt"   # nom du dossier dans episodes/
AUDIO_FILE   = "episode.webm"           # nom du fichier dans audio/

# Paramètres Whisper (optionnel)
MODEL_SIZE   = "base"   # tiny | base | small | medium | large-v3
DEVICE       = "cpu"    # cpu | cuda
# ─────────────────────────────────────────────────────────────────
```

```python
from faster_whisper import WhisperModel
from pathlib import Path

AUDIO_PATH  = Path(f"../audio/{AUDIO_FILE}")
OUTPUT_PATH = Path(f"../episodes/{EPISODE_SLUG}/transcript.txt")

if not AUDIO_PATH.exists():
    raise FileNotFoundError(f"Fichier audio introuvable : {AUDIO_PATH}")
if not OUTPUT_PATH.parent.exists():
    raise FileNotFoundError(f"Dossier épisode introuvable : {OUTPUT_PATH.parent}")

print(f"Transcription de : {AUDIO_PATH}")
print(f"Sortie vers      : {OUTPUT_PATH}")

model = WhisperModel(MODEL_SIZE, device=DEVICE, compute_type="int8")
segments, info = model.transcribe(str(AUDIO_PATH))

with open(OUTPUT_PATH, "w", encoding="utf-8") as f:
    for segment in segments:
        f.write(segment.text + "\n")

print("Done!")
```
